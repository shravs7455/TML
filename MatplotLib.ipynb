```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph 1')
```

```python
y=x*x
```

```python
plt.plot(x,y,'r*',linestyle=':',linewidth=2,markersize=20)
```

```python
plt.subplot(2,2,1) # number of graphs (2X2) and position of graph (1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'b*:')
plt.subplot(2,2,4)
plt.plot(x,y,'y*-')
```

```python
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.title("Sine Wave")
```
